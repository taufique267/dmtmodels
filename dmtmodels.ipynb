```python
import pandas as pd
import numpy as np
import spacy
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/twitter/twitter_training.csv')
df_test = pd.read_csv('/content/drive/MyDrive/twitter/twitter_validation.csv')

# Renaming columns
df_train.rename(columns={'2401': 'id',
                         'Borderlands': 'country',
                         'Positive': 'Label',
                         'im getting on borderlands and i will murder you all ,': 'text'}, inplace=True)

df_test.rename(columns={'3364': 'id',
                        'Facebook': 'country',
                        'Irrelevant': 'Label',
                        'I mentioned on Facebook that I was struggling for motivation to go for a run the other day, which has been translated by Tom‚Äôs great auntie as ‚ÄòHayley can‚Äôt get out of bed‚Äô and told to his grandma, who now thinks I‚Äôm a lazy, terrible person ü§£': 'text'}, inplace=True)

# Dropping NaN values
df_train.dropna(inplace=True)
df_test.dropna(inplace=True)

# Preprocessing function
nlp = spacy.load("en_core_web_sm")
def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    return " ".join(filtered_tokens)

# Applying preprocessing
df_train['Preprocessed Text'] = df_train['text'].apply(preprocess)
df_test['Preprocessed Text'] = df_test['text'].apply(preprocess)

# Splitting data into features and labels
X_train = df_train['Preprocessed Text']
y_train = df_train.Label

X_test = df_test['Preprocessed Text']
y_test = df_test.Label

# Encoding labels
le_model = LabelEncoder()
df_train['Label'] = le_model.fit_transform(df_train['Label'])
df_test['Label'] = le_model.fit_transform(df_test['Label'])
```

```python
# Vectorizing text data
v = TfidfVectorizer()
X_train_cv = v.fit_transform(X_train)
X_test_cv = v.transform(X_test)
```

```python
# Encoding string labels to integer labels
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)

# Initialize XGBoost Classifier
xgb_model = xgb.XGBClassifier()

# Train XGBoost model
xgb_model.fit(X_train_cv, y_train_encoded)
```

```python
# Encoding test labels if not already encoded
y_test_encoded = le.transform(y_test)  # Assuming 'le' is the LabelEncoder used to encode the labels

# Make predictions
y_pred = xgb_model.predict(X_test_cv)

# Evaluate model
print("Accuracy:", accuracy_score(y_test_encoded, y_pred))
print("Classification Report:\n", classification_report(y_test_encoded, y_pred))
```

**BY USING APPROXIMATE NEAREST NEIGHBOUR(ANN) ALGORITHM**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score, classification_report

# Load and preprocess training data
df_train = pd.read_csv('/content/drive/MyDrive/twitter/twitter_training.csv')
df_train.rename(columns={'2401': 'id',
                         'Positive': 'Sentiment',
                         'im getting on borderlands and i will murder you all ,': 'text'}, inplace=True)
df_train.dropna(inplace=True)

# Load and preprocess test data
df_test = pd.read_csv('/content/drive/MyDrive/twitter/twitter_validation.csv')
df_test.rename(columns={'3364': 'id',
                        'Irrelevant': 'Sentiment',
                        'I mentioned on Facebook that I was struggling for motivation to go for a run the other day, which has been translated by Tom‚Äôs great auntie as ‚ÄòHayley can‚Äôt get out of bed‚Äô and told to his grandma, who now thinks I‚Äôm a lazy, terrible person ü§£': 'text'}, inplace=True)
```

```python
# Splitting data
X_train = df_train['text']
y_train = df_train['Sentiment']
X_test = df_test['text']
y_test = df_test['Sentiment']
```

```python
# Vectorize text data
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
# Initialize and fit Nearest Neighbors model
ann_model = NearestNeighbors(n_neighbors=1, algorithm='auto')
ann_model.fit(tfidf_train)
```

```python
# Find nearest neighbor for each test sample
distances, indices = ann_model.kneighbors(tfidf_test)

# Map indices to sentiment labels
y_pred_indices = [y_train.iloc[idx] for idx in indices.flatten()]

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_indices)
print(f'Accuracy: {round(accuracy * 100, 2)}%')
```

```python
# Display classification report
print('Classification Report:')
print(classification_report(y_test, y_pred_indices))
```
